```python
#1.Discuss string slicing and provide examples
```

```python
#String slicing is a powerful feature in Python that allows you to extract a portion of a string by specifying a range of indices. The syntax for slicing is:

#code
#string[start:stop:step]
#Here’s what each part means:

#start: The starting index of the slice (inclusive). If omitted, it defaults to 0.
#stop: The ending index of the slice (exclusive). The slice includes characters up to, but not including, this index. If omitted, it defaults to the length of the string.
#step: The step size, which determines how many characters to skip between each in the slice. If omitted, it defaults to 1.
#Examples of String Slicing
#Basic Slicing:

#code
text = "Hello, World!"
sliced_text = text[0:5]
print(sliced_text)  # Output: Hello
#This example extracts the first five characters from the string.

#Omitting Start and Stop:

#code
text = "Hello, World!"
sliced_text = text[:5]  # Equivalent to text[0:5]
print(sliced_text)  # Output: Hello

sliced_text = text[7:]  # Equivalent to text[7:len(text)]
print(sliced_text)  # Output: World!
#Here, omitting start starts the slice from the beginning, and omitting stop slices until the end of the string.

#Using Step:

#code
text = "Hello, World!"
sliced_text = text[::2]
print(sliced_text)  # Output: Hlo ol!
#The step size of 2 means every second character is taken, effectively skipping one character each time.

#Negative Indices:

#code
text = "Hello, World!"
sliced_text = text[-6:-1]
print(sliced_text)  # Output: World
#Negative indices count from the end of the string, so -1 is the last character, -2 is the second to last, and so on.

#Reversing a String:

#code
text = "Hello, World!"
reversed_text = text[::-1]
print(reversed_text)  # Output: !dlroW ,olleH
#A step of -1 slices the string in reverse order.

#Summary
#String slicing in Python is a flexible way to access substrings using indices. By adjusting the start, stop, and step parameters, you can manipulate strings in various ways, such as extracting substrings, skipping characters, or even reversing the string.
```

```python
#Lists are one of the most versatile and widely used data structures in Python. They allow you to store an ordered collection of items, which can be of different types. Here are the key features of lists in Python:

#1. Ordered Collection
#Lists maintain the order of elements, which means the items have a defined sequence that will not change unless explicitly altered.
#Example:
#code
my_list = [1, 2, 3, 4]
print(my_list)  # Output: [1, 2, 3, 4]
#2. Mutable
#Lists are mutable, meaning that you can change, add, or remove items after the list has been created.
#Example:
#code
my_list = [1, 2, 3]
my_list[1] = 20
print(my_list)  # Output: [1, 20, 3]
#3. Can Contain Different Data Types
#A list can store elements of different data types, such as integers, strings, and even other lists.
#Example:
#code
my_list = [1, "Hello", 3.14, [10, 20]]
print(my_list)  # Output: [1, "Hello", 3.14, [10, 20]]
#4. Dynamic Size
#Lists in Python can grow and shrink in size as needed. You can append elements to the list or remove them dynamically.
#Example:
#code
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

my_list.pop(1)
print(my_list)  # Output: [1, 3, 4]
#5. Indexing and Slicing
#Lists support indexing, allowing you to access individual elements using their position in the list. Slicing enables you to access a sublist.
#Example:
#code
my_list = [10, 20, 30, 40, 50]
print(my_list[2])    # Output: 30
print(my_list[1:4])  # Output: [20, 30, 40]
#6. List Comprehensions
#Python provides a concise way to create lists using list comprehensions, which can be used to generate new lists by applying an expression to each item in an existing list.
#Example:
#code
squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]
#7. Methods for List Operations
#Python lists come with a variety of built-in methods for common operations like appending, removing, sorting, and reversing.
#Example:
#code
my_list = [3, 1, 4, 1, 5]
my_list.sort()
print(my_list)  # Output: [1, 1, 3, 4, 5]

my_list.reverse()
print(my_list)  # Output: [5, 4, 3, 1, 1]
#8. Nested Lists
#Lists can contain other lists as elements, creating nested or multi-dimensional lists.
#Example:
#code
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2])  # Output: 6
#9. Iterating Over Lists
#You can easily loop through a list to access each element.
#Example:
#code
my_list = [10, 20, 30]
for item in my_list:
    print(item)
# Output:
# 10
# 20
# 30
#10. Combining Lists
#Lists can be combined or concatenated using the + operator, and repeated using the * operator.
#Example:
#code
list1 = [1, 2]
list2 = [3, 4]
combined = list1 + list2
print(combined)  # Output: [1, 2, 3, 4]

repeated = list1 * 3
print(repeated)  # Output: [1, 2, 1, 2, 1, 2]
#Summary
#Lists in Python are flexible, allowing you to store, modify, and access data efficiently. Their mutability, support for various data types, and powerful methods make them an essential tool in Python programming.
```

```python
#In Python, lists are mutable, meaning you can access, modify, and delete elements within them. Here’s how you can perform each of these operations:

#1. Accessing Elements
#You can access elements in a list by using their index. Python uses zero-based indexing, so the first element has an index of 0.

#Example:
#code
# List of integers
numbers = [10, 20, 30, 40, 50]

# Accessing the first element
print(numbers[0])  # Output: 10

# Accessing the third element
print(numbers[2])  # Output: 30

# Accessing the last element using negative indexing
print(numbers[-1])  # Output: 50
#2. Modifying Elements
#Since lists are mutable, you can modify an element at a specific index by assigning a new value to that index.

#Example:
#code
# List of strings
fruits = ["apple", "banana", "cherry"]

# Modifying the second element
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

# Modifying the last element
fruits[-1] = "date"
print(fruits)  # Output: ['apple', 'blueberry', 'date']
#3. Deleting Elements
#You can delete elements from a list in several ways:

#Using del Statement: Removes the element at a specified index.
#Using pop() Method: Removes and returns the element at a specified index (or the last element if no index is provided).
#Using remove() Method: Removes the first occurrence of a specified value.
#Examples:
#Using del Statement:

#code
# List of integers
numbers = [10, 20, 30, 40, 50]

# Deleting the second element
del numbers[1]
print(numbers)  # Output: [10, 30, 40, 50]

# Deleting the last element
del numbers[-1]
print(numbers)  # Output: [10, 30, 40]
#Using pop() Method:

#code
# List of strings
fruits = ["apple", "banana", "cherry"]

# Popping the second element
removed_fruit = fruits.pop(1)
print(removed_fruit)  # Output: banana
print(fruits)         # Output: ['apple', 'cherry']

# Popping the last element
last_fruit = fruits.pop()
print(last_fruit)      # Output: cherry
print(fruits)          # Output: ['apple']
#Using remove() Method:

#code
# List of integers with duplicates
numbers = [10, 20, 30, 20, 40, 50]

# Removing the first occurrence of 20
numbers.remove(20)
print(numbers)  # Output: [10, 30, 20, 40, 50]

# Removing another occurrence of 20
numbers.remove(20)
print(numbers)  # Output: [10, 30, 40, 50]
#Summary
#Accessing elements: Use indexing to retrieve specific items from the list.
#Modifying elements: Assign a new value to an element using its index.
#Deleting elements: Use del for deletion by index, pop() to remove and retrieve an element, and remove() to delete the first occurrence of a value.
#These operations make Python lists highly flexible and easy to work with in a variety of scenarios.
```

```python
#Tuples and lists are both sequence data types in Python that can store collections of items. However, they have key differences in terms of mutability, syntax, and use cases. Below is a comparison of tuples and lists, along with examples to illustrate their differences and similarities.

#1. Mutability
#Lists: Lists are mutable, meaning you can modify them after their creation. You can change, add, or remove elements from a list.
#Tuples: Tuples are immutable, meaning once they are created, their elements cannot be changed, added, or removed.
#Example:

#code
# List example
my_list = [1, 2, 3]
my_list[1] = 20  # Modifying an element
print(my_list)  # Output: [1, 20, 3]

# Tuple example
my_tuple = (1, 2, 3)
# my_tuple[1] = 20  # This would raise a TypeError
#2. Syntax
#Lists: Lists are created using square brackets [].
#Tuples: Tuples are created using parentheses ().
#Example:

#code
# List
my_list = [1, 2, 3]

# Tuple
my_tuple = (1, 2, 3)
#3. Use Cases
#Lists: Lists are typically used when you need a collection of items that may change throughout the program. They are ideal for scenarios where you need to frequently modify, sort, or rearrange data.
#Tuples: Tuples are used when you want a collection of items that should not change after creation. They are often used to represent fixed collections of data, such as coordinates, RGB values, or database records.
#Example:

#code
# List use case: Managing a list of items in a shopping cart
shopping_cart = ["apple", "banana", "cherry"]
shopping_cart.append("date")
print(shopping_cart)  # Output: ['apple', 'banana', 'cherry', 'date']

# Tuple use case: Storing coordinates
coordinates = (40.7128, -74.0060)
print(coordinates)  # Output: (40.7128, -74.0060)
#4. Performance
#Lists: Lists are slightly slower than tuples because they are mutable and require more memory to accommodate changes.
#Tuples: Tuples are faster and more memory-efficient due to their immutability.
#Example:

#code
import timeit

# Performance comparison
list_time = timeit.timeit(stmt="[1, 2, 3, 4, 5]", number=1000000)
tuple_time = timeit.timeit(stmt="(1, 2, 3, 4, 5)", number=1000000)

print(f"List creation time: {list_time}")
print(f"Tuple creation time: {tuple_time}")
#5. Functions and Methods
#Lists: Lists have a wide variety of built-in methods like append(), remove(), sort(), reverse(), etc., which allow for extensive manipulation.
#Tuples: Tuples have fewer built-in methods, mainly count() and index(), because they are intended to be immutable.
#Example:

#code
# List methods
my_list = [3, 1, 4]
my_list.append(2)
my_list.sort()
print(my_list)  # Output: [1, 2, 3, 4]

# Tuple methods
my_tuple = (3, 1, 4, 1)
print(my_tuple.count(1))  # Output: 2
print(my_tuple.index(4))  # Output: 2
#6. Memory Usage
#Lists: Because they are mutable, lists use more memory than tuples.
#Tuples: Tuples, being immutable, are more memory-efficient than lists.
#Example:

#code
import sys

my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

print(sys.getsizeof(my_list))  # Example output: 104
print(sys.getsizeof(my_tuple))  # Example output: 88
#7. Nested Structures
#Lists: Lists can contain other lists, making them suitable for creating multi-dimensional arrays or more complex data structures.
#Tuples: Tuples can also contain other tuples, but they cannot be modified once created.
#Example:

#code
# Nested list
nested_list = [[1, 2], [3, 4]]
nested_list[0][1] = 20
print(nested_list)  # Output: [[1, 20], [3, 4]]

# Nested tuple
nested_tuple = ((1, 2), (3, 4))
# nested_tuple[0][1] = 20  # This would raise a TypeError
#Summary
#Mutability: Lists are mutable; tuples are immutable.
#Syntax: Lists use square brackets []; tuples use parentheses ().
#Use Cases: Lists are for collections that may change; tuples are for fixed collections.
#Performance: Tuples are generally faster and use less memory than lists.
#Methods: Lists have more methods available for manipulation than tuples.
#Memory Usage: Tuples are more memory-efficient.
#Nested Structures: Both can be nested, but lists are more flexible.
#Understanding these differences helps in choosing the appropriate data structure based on the requirements of your program.
```

```python
#Sets are an important data structure in Python, designed to store unique elements without any particular order. They are highly useful when you need to perform operations like union, intersection, and difference on collections of items. Here are the key features of sets in Python along with examples:

#1. Unordered Collection
#Sets are unordered, meaning that the elements do not have a specific order, and you cannot access elements by index.
#Example:
#code
my_set = {3, 1, 4, 2}
print(my_set)  # Output could be {1, 2, 3, 4}, order is not guaranteed
#2. Unique Elements
#Sets automatically remove duplicates. If you add an element that already exists in the set, it will not be added again.
#Example:
#code
my_set = {1, 2, 3, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}
#3. Mutable
#Sets are mutable, meaning you can add, remove, or update elements. However, the elements themselves must be immutable (e.g., numbers, strings, tuples).
#Example:
#code
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}
#4. No Indexing or Slicing
#Since sets are unordered, they do not support indexing, slicing, or other sequence-like behavior.
#Example:
#code
my_set = {1, 2, 3}
# print(my_set[0])  # This would raise a TypeError
#5. Efficient Membership Testing
#Sets are optimized for checking whether an element is a member of the set. This operation is faster in sets compared to lists.
#Example:
#code
my_set = {1, 2, 3, 4}
print(3 in my_set)  # Output: True
print(5 in my_set)  # Output: False
#6. Set Operations
#Sets support mathematical operations like union, intersection, difference, and symmetric difference.

#Union (| or union()): Combines elements from two sets.

#Intersection (& or intersection()): Returns only elements found in both sets.

#Difference (- or difference()): Returns elements in the first set but not in the second.

#Symmetric Difference (^ or symmetric_difference()): Returns elements in either set, but not in both.

#Examples:

#code
set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Union
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5}

# Intersection
print(set_a & set_b)  # Output: {3}

# Difference
print(set_a - set_b)  # Output: {1, 2}

# Symmetric Difference
print(set_a ^ set_b)  # Output: {1, 2, 4, 5}
#7. Set Comprehensions
#Similar to list comprehensions, Python supports set comprehensions for generating sets based on existing sequences or conditions.
#Example:
#code
squared_set = {x**2 for x in range(5)}
print(squared_set)  # Output: {0, 1, 4, 9, 16}
#8. Immutability with frozenset
#Python provides a frozenset type, which is an immutable version of a set. Once created, a frozenset cannot be modified.
#Example:
#code
my_frozenset = frozenset([1, 2, 3])
# my_frozenset.add(4)  # This would raise an AttributeError
print(my_frozenset)  # Output: frozenset({1, 2, 3})
#9. Duplicate Elimination
#One common use of sets is to remove duplicates from a list or other iterable.
#Example:
#code
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3, 4, 5}
#10. Combining Sets
#You can combine sets using operations like union and update.
#Example:
#code
set_a = {1, 2}
set_b = {3, 4}
set_a.update(set_b)
print(set_a)  # Output: {1, 2, 3, 4}
#Summary
#Unordered and Unique: Sets store elements without order and automatically ensure all elements are unique.
#Mutable: Elements can be added or removed from a set, but the elements themselves must be immutable.
#No Indexing: You cannot access elements by index since sets are unordered.
#Efficient Membership Testing: Checking for an element's presence is fast and efficient.
#Mathematical Operations: Supports operations like union, intersection, difference, and symmetric difference.
#Set Comprehensions: Provides a concise way to create sets using expressions.
#Immutability with frozenset: Provides an immutable set type when needed.
#Sets are particularly useful for situations where you need to manage collections of unique elements and perform operations that are common in mathematical set theory.
```

```python
#Tuples and sets are both important data structures in Python, each with its unique characteristics and use cases. Here’s a detailed discussion of when and why you might choose to use tuples or sets in Python programming:

#Use Cases of Tuples
#1. Immutable Collections
#When immutability is required: Tuples are immutable, making them ideal for storing data that should not change throughout the program. This immutability ensures that the data remains consistent and unmodified.
#Example: Storing constants, such as configuration settings or fixed options.
#code
connection_params = ("localhost", 8080)
#2. Return Multiple Values from Functions
#Returning multiple values: Functions in Python can return multiple values using tuples. This is a common use case when a function needs to provide more than one piece of related information.
#Example: Returning both the quotient and remainder from a division operation.
#code
def divide(x, y):
    quotient = x // y
    remainder = x % y
    return quotient, remainder

result = divide(10, 3)
print(result)  # Output: (3, 1)
#3. Structured Data
#Structured data representation: Tuples can be used to group related pieces of data together. This is useful in scenarios where you need to represent a record with a fixed structure.
#Example: Storing a point in 2D or 3D space, where the coordinates are logically grouped together.
#code
point = (3, 4)
color = (255, 0, 0)  # RGB value for red
#4. Using Tuples as Dictionary Keys
#Dictionary keys: Because tuples are immutable, they can be used as keys in dictionaries (unlike lists). This is particularly useful when you need a composite key.
#Example: Using a tuple of coordinates as a dictionary key.
#code
locations = {(40.7128, -74.0060): "New York", (34.0522, -118.2437): "Los Angeles"}
print(locations[(40.7128, -74.0060)])  # Output: New York
#5. Iteration and Unpacking
#Tuple unpacking: Tuples support unpacking, which can make code more readable and concise, especially when dealing with functions that return multiple values.
#Example: Looping through a list of tuples and unpacking each tuple into separate variables.
#code
people = [("Alice", 30), ("Bob", 25), ("Charlie", 35)]
for name, age in people:
    print(f"{name} is {age} years old")
#6. Performance
#Performance optimization: Tuples are more memory-efficient than lists and can be faster in scenarios where a collection is created and iterated over without modification.
#Example: Using tuples in scenarios where a large number of small, immutable collections are created.
#code
records = [("John", 22), ("Jane", 28), ("Jim", 31)]
#Use Cases of Sets
#1. Removing Duplicates
#Unique elements: Sets are ideal when you need to store a collection of unique elements, automatically eliminating duplicates.
#Example: Removing duplicates from a list.
#code
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)
print(unique_numbers)  # Output: {1, 2, 3, 4, 5}
#2. Mathematical Set Operations
#Set theory operations: Sets are designed to efficiently perform mathematical set operations like union, intersection, difference, and symmetric difference. These operations are useful in a variety of applications, such as data comparison and filtering.
#Example: Finding common elements in two lists.
#code
set_a = {1, 2, 3}
set_b = {3, 4, 5}
common_elements = set_a & set_b
print(common_elements)  # Output: {3}
#3. Membership Testing
#Efficient membership testing: Sets offer O(1) average time complexity for membership checks, making them ideal for scenarios where you need to frequently check the presence of elements.
#Example: Checking if an element exists in a large dataset.
#code
allowed_ips = {"192.168.1.1", "192.168.1.2", "192.168.1.3"}
if "192.168.1.4" in allowed_ips:
    print("IP is allowed")
else:
    print("IP is not allowed")
#4. Filtering Data
#Data filtering: Sets can be used to filter out unwanted elements from a collection, or to retain only elements that meet certain criteria.
#Example: Filtering out elements from one list that are present in another.
#code
list_a = [1, 2, 3, 4]
list_b = [3, 4, 5, 6]
filtered_list = list(set(list_a) - set(list_b))
print(filtered_list)  # Output: [1, 2]
#5. Storing Unordered Collections
#Unordered collections: Use sets when the order of elements doesn’t matter, but uniqueness does.
#Example: Managing a collection of tags where order is irrelevant, but duplicate tags should not exist.
#code
tags = {"python", "coding", "development", "python"}
print(tags)  # Output: {'python', 'coding', 'development'}
#6. Handling Large Data Sets
#Large data handling: When dealing with large datasets, especially when you need to perform operations like intersection or difference, sets are more efficient than other data structures.
#Example: Finding common customers between two large databases.
#code
customers_2023 = {"Alice", "Bob", "Charlie", "Diana"}
customers_2024 = {"Charlie", "Diana", "Edward", "Fiona"}
common_customers = customers_2023.intersection(customers_2024)
print(common_customers)  # Output: {'Charlie', 'Diana'}
#Summary
#Tuples: Use tuples when you need a fixed, immutable collection of elements, such as when returning multiple values from a function, using as dictionary keys, or representing structured data that should not change.
#Sets: Use sets when you need to manage collections of unique elements, perform mathematical set operations, efficiently check for membership, or filter data, especially when dealing with large and unordered data.
#Understanding the strengths of tuples and sets will help you choose the appropriate data structure for your specific use case, leading to more efficient and readable code.
```

```python
#Dictionaries in Python are mutable collections that store data in key-value pairs. They are highly versatile and allow for quick access, modification, and management of data. Here’s a detailed explanation of how to add, modify, and delete items in a dictionary, along with examples.

#1. Adding Items to a Dictionary
#Adding a New Key-Value Pair
#To add a new key-value pair to a dictionary, simply assign a value to a new key.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 25}
my_dict['city'] = 'New York'  # Adding a new key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
#Using update() Method
#You can also use the update() method to add multiple key-value pairs at once. If the keys already exist, their values will be updated.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 25}
my_dict.update({'city': 'New York', 'email': 'alice@example.com'})
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York', 'email': 'alice@example.com'}
#2. Modifying Items in a Dictionary
#Updating the Value of an Existing Key
#To modify the value associated with an existing key, assign a new value to that key.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
my_dict['age'] = 26  # Modifying the value of the 'age' key
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}
#Using update() Method
#The update() method can also be used to modify existing keys.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
my_dict.update({'age': 26, 'city': 'Los Angeles'})
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'}
#3. Deleting Items from a Dictionary
#Using del Statement
#The del statement can be used to remove a key-value pair from a dictionary by specifying the key.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 26, 'city': 'New York'}
del my_dict['age']  # Deleting the 'age' key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York'}
#Using pop() Method
#The pop() method removes a key from the dictionary and returns its value. If the key is not found, an optional default value can be returned; otherwise, a KeyError is raised.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 26, 'city': 'New York'}
age = my_dict.pop('age')  # Removes 'age' and returns its value
print(age)      # Output: 26
print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York'}
#Using popitem() Method
#The popitem() method removes and returns the last key-value pair as a tuple. It is useful in scenarios where the dictionary is being treated as a stack (LIFO structure).
#Example:
#code
my_dict = {'name': 'Alice', 'age': 26, 'city': 'New York'}
last_item = my_dict.popitem()  # Removes and returns the last key-value pair
print(last_item)  # Output: ('city', 'New York')
print(my_dict)    # Output: {'name': 'Alice', 'age': 26'}
#Using clear() Method
#The clear() method removes all items from the dictionary, leaving it empty.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 26, 'city': 'New York'}
my_dict.clear()  # Removes all key-value pairs
print(my_dict)   # Output: {}
#Summary
#Adding Items: You can add items by assigning a value to a new key or using the update() method.
#Modifying Items: Modify items by reassigning a new value to an existing key or using update() for multiple changes.
#Deleting Items: Delete items using del, pop(), popitem(), or clear() methods, depending on whether you want to remove specific keys, the last item, or all items in the dictionary.
#These operations make dictionaries a flexible and powerful tool for managing dynamic collections of key-value pairs in Python.
```

```python

#In Python, dictionary keys must be immutable types. This immutability requirement is crucial for ensuring that dictionaries can maintain their integrity and consistency during operations. Here’s a detailed discussion on why dictionary keys need to be immutable, along with examples:

#Importance of Dictionary Keys Being Immutable
#Hash Consistency

#Hashing Requirement: Dictionary keys are hashed to determine their location in memory. The hash value is used for quick look-up, insertion, and deletion of key-value pairs. If a key were mutable, its hash value could change after insertion, causing the dictionary to become inconsistent.
#Example:
#code
my_dict = {}
key = (1, 2, 3)  # Change the list to a tuple
my_dict[key] = 'value'  # Raises TypeError
#In this example, trying to use a list as a key will result in a TypeError because lists are mutable and do not have a fixed hash value.
#Data Integrity

#Consistency: Immutable keys ensure that once a key-value pair is added to a dictionary, the key will not change, thereby maintaining the consistency of the data. This immutability guarantees that the dictionary's internal data structures remain valid and reliable.
#Example:
#code
my_dict = {'name': 'Alice', 'age': 25}
# If 'name' was mutable, its value could change, potentially disrupting the dictionary's structure
#Avoiding Unexpected Behavior

#Predictable Behavior: Immutable keys prevent unexpected behavior that might occur if the key could be modified after it was used in the dictionary. This ensures that operations such as look-ups and deletions behave consistently.
#Example:
#code
my_dict = {(1, 2): 'value'}  # Tuple (immutable) as a key
print(my_dict[(1, 2)])  # Output: 'value'

# The following would fail with mutable types:
my_dict = {(1, 2): 'value'}  # Use a tuple (immutable) as a key
#Efficiency

#Performance: Using immutable keys allows dictionaries to be highly efficient in terms of performance. Since the hash value of a key is constant, dictionary operations like insertion, deletion, and lookup can be performed quickly.
#Example:
#code
my_dict = {1: 'one', 2: 'two'}
# Using immutable integers as keys ensures efficient look-ups and insertions
#Examples of Immutable and Mutable Types
#Immutable Types as Keys
#Examples: int, float, str, tuple (provided the tuple itself contains only immutable elements)
#Example:
#code
my_dict = {1: 'one', 2.5: 'two point five', 'key': 'value', (1, 2): 'tuple key'}
print(my_dict[1])        # Output: 'one'
print(my_dict[2.5])      # Output: 'two point five'
print(my_dict['key'])    # Output: 'value'
print(my_dict[(1, 2)])  # Output: 'tuple key'
#Mutable Types as Keys (Not Allowed)
#Examples: list, dict, set
#Example:
#code
# This will raise TypeError because lists are mutable
my_dict = {}
key_tuple = (1, 2, 3)  # Use a tuple instead of a list
my_dict[key_tuple] = 'value'  # This will work now
print(my_dict)
#Summary
#Hashing: Immutable keys ensure that hash values remain constant, which is essential for dictionary operations.
#Data Integrity: Immutable keys preserve the consistency of dictionary data, preventing unexpected modifications.
#Predictable Behavior: Immutable keys ensure that dictionary operations like look-ups and deletions are predictable and reliable.
#Efficiency: Immutable keys contribute to the performance efficiency of dictionary operations by maintaining a stable hash value.
#In summary, the immutability of dictionary keys is a fundamental requirement for the integrity, efficiency, and reliability of Python dictionaries. It allows for fast access and manipulation of data while preventing issues that could arise from key modifications.
```
